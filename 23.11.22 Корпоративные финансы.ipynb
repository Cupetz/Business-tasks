```python
def NVP(CF, r, n, IC):
    sum_nvp = 0
    for i in range(1, n+1):
        period_nvp = CF[i-1]/(1+r)**i
        print(f'Период {i} : {period_nvp}')
        sum_nvp += period_nvp
    return sum_nvp - IC
```

```python
print(f'Итоговый доход : {NVP([1000_000, 1500_000, 1400_000, 1500_000, 1200_000], 0.2, 5, 3000_000)}')
```

```python
def PI(CF, r, n, IC):
    sum_nvp = 0
    for i in range(1, n+1):
        period_nvp = CF[i-1]/(1+r)**i
        sum_nvp += period_nvp
    return sum_nvp/IC
```

```python
print(f'Индекс доходности : {PI([1000_000, 1500_000, 1400_000, 1500_000, 1200_000], 0.2, 5, 3000_000)}')
```

```python
def PBP(CF, r, n, IC):
    
    current_ic = -IC
    year = None
    
    all_nvp = []
    all_ic = []
    
    for i in range(1, n+1):
        
        period_nvp = CF[i-1]/(1+r)**i
        current_ic += period_nvp
        
        all_nvp.append(period_nvp)
        all_ic.append(current_ic)
        
        print(f'Текущий выход на прибыль в период {i} : {round(current_ic, 3)}')
        print(f'Текущий дисконтированный доход в период {i} : {round(period_nvp, 3)}')
        
        if year == None and current_ic > 0:
            year = i
            
    print(f'Год выхода в прибыль : {year-1}')
    ic = abs(all_ic[year-2])
    nvp = all_nvp[year-1]
    print(f'ic : {round(ic, 3)}')
    print(f'nvp : {round(nvp, 3)}')
        
    return year-1 + (ic/nvp)
```

```python
print(f'Дисконтированный срок окупаемости : {round(PBP([1000_000, 1500_000, 1400_000, 1500_000, 1200_000], 0.2, 5, 3000_000), 3)}')
```
